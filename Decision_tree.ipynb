```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data=pd.read_csv("/content/gdrive/MyDrive/DSMl/drug200.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
input=data.drop("Drug",axis=1)
```

```python
input
```

```python
data1=data.replace({'Drug':{'drugY':0,'drugX':1,'drugA':2,'drugB':3,'drugC':4}})
```

```python
target=data1['Drug']
target.head()
```

```python
data.dtypes
```

```python
sex_n=LabelEncoder()
BP_n=LabelEncoder()
Cholesterol_n=LabelEncoder()
```

```python
input['sex_n']=sex_n.fit_transform(input['Sex'])
```

```python
input['Bp_n']=BP_n.fit_transform(input['BP'])
```

```python
input['Cholesterol_n']=Cholesterol_n.fit_transform(input['Cholesterol'])
```

```python
input.head()
```

```python
X=input.drop(['Sex','BP','Cholesterol'],axis=1)
```

```python
X
```

```python
Y=target
Y.head()
```

```python
model=tree.DecisionTreeClassifier()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
print('X_train',X_train.shape)
print('X_test',X_test.shape)
print('Y_train',Y_train.shape)
print('Y_test',Y_test.shape)
```

```python
model.fit(X,Y)
```

```python
prediction=model.predict(X_test)
```

```python
print(" The accuracy of DecissionTreeClassifier",metrics.accuracy_score(prediction,Y_test))
```
