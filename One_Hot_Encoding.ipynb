```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error as mae
```

```python
data=pd.read_csv("carprices.csv")
```

```python
data
```

One Hot Encoding

```python
data.rename(columns={'car Model': 'car_model'}, inplace=True)
data
```

```python
data
```

```python
dummies=pd.get_dummies(data['Car Model'])

```

```python
dummies
```

```python
merged=pd.concat([data,dummies],axis='columns')
merged
```

```python
final=merged.drop(['Car Model'],axis=1)
```

```python
final
```

```python
final=final.drop(['Mercedez Benz C class'],axis=1)
```

```python
X=final.drop(['Sell Price($)'],axis=1)
```

```python
X.head()
```

```python
Y=final.Mileage
```

```python
Y.head()
```

```python
model=LinearRegression()
```

```python
model.fit(X,Y)
```

1. Predict the price of a Mercedez Benz that is 4 yr old with a mileage of 45000


```python
price=model.predict([[45000,4,0,0]])
print("price",price)
```

2. Predict the price of a BMW X5 that is 7 years old with a mileage of 86000


```python
price=model.predict([[86000,7,0,1]])
print("price",price)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print('X_train',X_train.shape)
print('X_test',X_test.shape)
print('Y_train',Y_train.shape)
print('Y_test',Y_test.shape)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train_prediction=model.predict(X_train)
```

```python
price_model=LinearRegression()
```

```python
price_model.fit(X_train,Y_train)
```

```python
train_prediction=price_model.predict(X_train)
```

```python
print(train_prediction)
```

```python
print("training error:",mae(Y_train,train_prediction))
```

```python
test_prediction=price_model.predict(X_test)
```

```python
print(test_prediction)
```

```python
print("validation error:",mae(Y_test,test_prediction))
```
